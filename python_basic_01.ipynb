## <p style="background-color:#9d4f8c; font-family:newtimeroman; color:#FFF9ED; font-size:175%; text-align:center; border-radius:10px 10px;">Some Important PEP 8 Rules</p>

<a id="6"></a>
<a href="#toc" class="btn btn-primary btn-sm" role="button" aria-pressed="true" 
style="color:blue; background-color:#dfa8e4" data-toggle="popover">Content</a>

PEP 8 ise Python kod yazım stilini açıklar. Bu dökümanda özetle aşağıdaki kod yazım kuralları önerilmektedir:
-Kod girintisi için 4 boşluk (1 TAB) bırakılmalıdır, 
• Bir kod satırı en fazla 79 (açıklama ve yorum satırları ise 72) karakter içermelidir, 
• Parantez içlerini kullanırken boşluk bırakılmamalıdır. Örneğin; 

```python
# Doğrusu: spam(ham[1], {eggs: 2}) 
# Yanlışı: spam( ham[ 1 ], { eggs: 2 } )
```

```python
#Virgül, noktalı virgül veya iki nokta üst üste işareti öncesi ile bitişik sonrası ile ayrık olmalıdır. Örneğin; 

# Doğrusu:
x = 5; y = 4
if x == 4: print(x, y); x, y = y, x 
# Yanlışı: 
if x == 4 : print(x , y) ; x , y = y , x
```

```python
# Bir operatörün her iki tarafında da eşit (birer adet) boşluk bırakılmalıdır. Örneğin;
# Doğrusu:
i = i + 1
# Yanlışı: 
i =i +1
```

```python
#Birden fazla kelimeden oluşan değişken isimlerinde _ kullanılması önerilir. Örneğin;
# Doğrusu:
ad_soyad
# Yanlışı: 
ad-soyad
```

```python
#Sabit (değeri değişmeyen değişken) tanımlamalarında büyük harf kullanılması önerilir. Örneğin;
# Doğrusu: 
PI = 3.14
# Yanlışı: 
Pi = 3.14
```

```python
# True / False ifadelerini ‘==’ operatörü ile doğrudan karşılaştırmayın. Örneğin;
# Doğrusu: 
if a:
# Yanlışı: 
if a == True:
```

Kod yazımında her zaman netlik ve sadelik önerilir. Örneğin;
# Doğrusu:
if foo == 'blah':
    do_blah_thing()

do_one()
do_two()

# Yanlışı: 
if foo == 'blah': do_blah_thing()
do_one(); do_two()

Python dilinde  açıklama(yorum) satırları için # simgesi kullanılır. 

Açıklama (yorum) satırlarını derleyiciler dikkate almaz. 

Bu satırlar, programcı için yazılan kod satırı ile ilgili hatırlatıcı, açıklayıcı bilgiler verir.

Eğer tek satırlık bir açıklama yapılacaksa ‘#’ karakteri,
